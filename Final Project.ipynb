**Metric**

Pada umumnya untuk regresi, kita bisa menggunakan metric seperti MAE,RMSE,R2,MAPE,dan MedAE.Pemilihan metric ini berdasarkan karakteristik datanya.Evaluasi metrik yang dapat digunakan jika memiliki banyak outlier adalah MedAE karena berdasarkan https://scikit-learn.org/stable/modules/model_evaluation.html#median-absolute-error:~:text=3.3.4.6.-,Median%20absolute%20error,-%C2%B6 metric ini robust terhadap outlier,RMSE(jika tidak ada outlier). Penggunaan MedAE ini robust terhadap outliers, dan untuk MedAE hasil errornya didapat dari median absolut errornya.Untuk mempermudah interpretasi MedAE berdasarkan https://help.pecan.ai/en/articles/6456388-model-performance-metrics-for-regression-models#:~:text=Percentage%20Error%20(RMSPE)-,Median%20Absolute%20Percentage%20Error%20(MdAPE),-Median%20Absolute%20Percentage kita dapat membuat custom metric yaitu MdAPE yang merupakan persentase MedAE, RMSE adalah nilai rata-rata error kuadrat yang setelah itu di akarkan agar lebih mudah untuk dipahami. Selain itu kita juga akan menggunakan R2 score yang dapat mengukur kemampuan model dalam menjelaskan target variabel menggunakan variabel dependennya.Semakin kecil nilai RMSE, dan MedAE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. Sebaliknya semakin besar nilai R-Squared yang memiliki nilai antara 0-1, maka semakin baik model tersebut.
Rumus:
- MedAE
$$\text{MedAE}(y, \hat{y}) = \text{median}(\mid y_1 - \hat{y}_1 \mid, \ldots, \mid y_n - \hat{y}_n \mid).$$
- RMSE
$$\text{RMSE}(y, \hat{y}) = \sqrt{\frac{\sum_{i=0}^{N - 1} (y_i - \hat{y}_i)^2}{N}}$$
- R2
$$R^2(y, \hat{y}) = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$$

```python
# Membuat boxplot untuk melihat outliers
columns_for_boxplot=[col for col in merged_df.columns if merged_df[col].dtypes!='object']

# Mengambil warna secara acak
colors = ['#'+''.join(random.choices('0123456789ABCDEF', k=6)) for _ in range(len(columns_for_boxplot))]

num_cols = 2
num_rows = 10
total_plots = num_cols * num_rows
num_plots = min(len(columns_for_boxplot), total_plots)

plt.figure(figsize=(15, 30))

for i in range(num_plots):
    plt.subplot(num_rows, num_cols, i+1)
    sb.boxplot(data=merged_df,x=columns_for_boxplot[i],color=colors[i])
    plt.title(columns_for_boxplot[i].capitalize())

plt.tight_layout()
plt.show()
```

* Dari boxplot di atas kita bisa melihat semua kolom numerikal memiliki outlier.
* Untuk outlier akan diatasi dengan robust scaler,namun kita akan cek terlebih dahulu apakah outlier tersebut merupakan data sebenarnya atau tidak.

```python
def show_outlier(col_name,table=True):
    Q1 = merged_df[col_name].quantile(0.25)
    Q3 = merged_df[col_name].quantile(0.75)
    IQR = Q3-Q1
    outlier_df = merged_df[(merged_df[col_name]<(Q1-1.5*IQR))|(merged_df[col_name]>(Q3+1.5*IQR))]
    if table==False:
        print(f'Kolom {col_name} memiliki {outlier_df.shape[0]} baris data outlier yang merupakan {round(outlier_df.shape[0]/merged_df.shape[0]*100,2)}% dari keseluruhan data')
    print(f'Data dianggap sebagai outlier apabila nilainya kurang dari {Q1-1.5*IQR} atau lebih dari {Q3+1.5*IQR} \n')
    if True:
        return outlier_df
```

```python
for col in [col for col in merged_df.columns if merged_df[col].dtypes!='object']:
    show_outlier(col,table=False)
```

* Seluruh kolom numerikal tidak berdistribusi normal, sehingga untuk imputasi akan digunakan median.

```python
def kruskal_test(col_test):
    for col in [col for col in merged_df.columns if merged_df[col].dtypes=='object']:
        group=[merged_df[(merged_df[col] == i)][col_test].values for i in merged_df[col].unique()]
        stat,pvalue=kruskal(*group)
        if pvalue<=0.05:
            print(f'Menolak H0. untuk setiap value {col} unik terdapat perbedaan {col_test} signifikan ')
        else:
            print(f'Gagal Menolak H0. Tidak terbukti untuk setiap value {col} unik terdapat perbedaan {col_test} signifikan')

kruskal_test('engineSize')
```

Disini median yang kita ambil hanya data yang memiliki engine size > 0.

```python
def anomaly_impute_engine(df):
    median_value = merged_df[(merged_df['model']==df['model'])&(merged_df['transmission']==df['transmission'])&
                  (merged_df['fuelType']==df['fuelType'])&(merged_df['engineSize']>0)]['engineSize'].median()
    if np.isnan(median_value):
      median_value = merged_df[(merged_df['model']==df['model'])&(merged_df['engineSize']>0)]['engineSize'].median() 
      return median_value
    else:
      return median_value
```

Konversi liter/100km ke mpg UK: https://www.inchcalculator.com/convert/liter-per-hundred-kilometers-to-mile-per-gallon-uk/

```python
conversion_factor = 282.481

def convert_mpg(mpg_value):
    if mpg_value < 10:
        return  conversion_factor/mpg_value
    else:
        return mpg_value
```

* Setelah dikonversikan ke mpg, perlu dicek lagi apakah hasil konversi tersebut sudah termasuk dalam rentang nilai mpg pada model tersebut.

```python
def anomaly_check_mpg(col,df):
    anomaly_index=[]
    for i in df.index:
        q_1=merged_df[(merged_df['model']==df.loc[i,'model'])][col].quantile(0.25)
        q_3=merged_df[(merged_df['model']==df.loc[i,'model'])][col].quantile(0.75)
        iqr = q_3-q_1
        if df.loc[i,col]<(q_1-1.5*iqr) or df.loc[i,col]>(q_3+1.5*iqr):
            anomaly_index.append(i)
    return df.loc[anomaly_index]
```

* Untuk kolom model dan brand, kita gunakan binary encoder karena jumlah nilai diskrit yang banyak.
* Metric yang akan digunakan adalah MedAE(robust terhadap outlier) dan R2 sebagai metric pendukung.

### Encoding & Scaling

```python
scaler=RobustScaler()
scaler_col=['year','mileage','mpg','engineSize']
oh=OneHotEncoder(drop='first')
oh_col=['transmission','fuelType','brand']

be=ce.BinaryEncoder()
be_col=['model']



transformer=ColumnTransformer([
    ('onehot',oh,oh_col),
    ('binary_encoding',be,be_col),
    ('robust_sc',scaler,scaler_col),

],remainder='passthrough')
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score,KFold,GridSearchCV,cross_validate
from lightgbm import LGBMRegressor
from sklearn.metrics import median_absolute_error, r2_score
import time
ridge = Ridge(random_state=42)
knn = KNeighborsRegressor()
xgb = XGBRegressor(random_state=42)
lgb = LGBMRegressor(random_state=42)
dt = DecisionTreeRegressor(random_state=42)
```

### Benchmark Models

```python
scoring = ['neg_median_absolute_error','r2']
models = [ridge,dt,knn,xgb,lgb]
medae_cv=[]
r2_cv=[]
runtime=[]
cv = KFold(n_splits=10,shuffle=True,random_state=42)
for model in models:
    estimator = Pipeline([('preprocessing',transformer),
                          ('model',model)])
    start_time = time.time()
    scores = cross_validate(estimator,x_train,y_train,scoring=scoring,cv=cv)
    medae_cv.append(scores['test_neg_median_absolute_error'].mean())
    r2_cv.append(scores['test_r2'].mean())
    runtime.append(time.time()-start_time)
def_performance = pd.DataFrame({'Model':models,'Runtime':runtime,'MedAE':medae_cv,'R2':r2_cv})
def_performance
```

* Berdasarkan hasil cross validasi, MedAE KNN memiliki performa paling baik, dan untuk R2 justru XGBoost lah yang paling baik.
* Jika dilihat dari runtime model KNN membutuhkan waktu kira-kira 2x lebih lama dari model XGBoost.
* Dengan pertimbangan kedepannya dimana data akan semakin banyak sehingga model KNN akan cenderung kurang dapat dipakai.Dan jika dilihat dari sisi performa sepertinya tidak terdapat perbedaan yang terlalu jauh antara KNN dan XGBoost, kemudian XGBoost juga menang pada metric R2.
* Untuk sekarang model yang akan kita coba tuning adalah KNN dan XGBoost
* XGBoost juga unggul dalam intepretasi model dibanding KNN.

* Untuk KNN tuning hanya dilakukan pada nilai n_neighbors saja.
  - n_neighbors - Mengatur berapa banyak data point terdekat yang akan digunakan untuk prediksi.
* Untuk XGBoost ada banyak hyperparameter yang akan dituning, berdasarkan penjelasan pada dokumentasinya https://xgboost.readthedocs.io/en/stable/parameter.html dan https://xgboost.readthedocs.io/en/stable/tutorials/param_tuning.html
  Kita akan tuning pada beberapa parameter yang dapat mengatur kompleksitas dan efek regularisasi model tersebut saja seperti:
  - n_estimators - Mengatur jumlah weak learner yang digunakan model, semakin banyak model semakin lambat dan semakin kompleks
  - max_depth - Mengatur kompleksitas kedalaman tiap pohon, value terlalu kecil berakibat underfitting sebaliknya terlalu besar overfitting
  - eta - Mengatur seberapa besar atau cepat model mengupdate weight tiap training step, value yang terlalu tinggi bisa mengakibatkan model tidak converge.
  - reg_alpha - regularisasi L1 untuk mengatasi overfitting(semakin besar semakin kuat efek regularisasinya)
  - reg_lambda - regularisasi L2 untuk mengatasi overfitting(semakin besar semakin kuat efek regularisasinya)
  - gamma - mengatur batas minimum improvement pada loss(case kita adalah MedAE) yang dibutuhkan pohon untuk melakukan pembagian lagi.
  - colsample_bytree - Mengatur jumlah kolom yang digunakan oleh tiap weak learner
  - subsample - Mengatur jumlah baris yang digunakan oleh tiap weak learner

#### Custome Metric

```python
def mdape(actual,predict):
    return np.median(abs((actual-predict)/actual))*100
```

### HyperParameter Tuning

#### KNN 

```python
knn.get_params()
```

* Default Value:
  - n_neighbors ==> 5

```python
params_knn = {
    "model__n_neighbors": list(np.arange(1,20))
    }

knn_pipe = Pipeline([('preprocessing',transformer),('model',KNeighborsRegressor())])
knn_search = GridSearchCV(knn_pipe, 
                            scoring='neg_median_absolute_error', 
                            param_grid=params_knn,
                            cv=cv, n_jobs=-1)
knn_search.fit(x_train, y_train)
print(knn_search.best_score_)
print(knn_search.best_params_)
```

* Setelah di tuning, terlihat tidak ada perbedaan yang terlalu besar pada MedAE sebelum dan sesudah tuning dimana MedAE sebelum tuning sebesar 771.84 dan setelah tuning sebesar 771.18.Walaupun demikian KNN terbaik adalah KNN setelah tuning.

* Terlihat untuk value MedAE pada test set hampir sama dengan best score dari gridsearchcv.
* Sehingga dapat dikatakan model KNN tuning ini memiliki generalisasi yang baik pada data yang belum pernah terlihat sebelumnya.
* Dari R2 yang dihasilkan, model dengan menggunakan dependen variabel mampu menjelaskan hampir 96% dari variabel target pada test set.

#### XGBoost

```python
params = {
    "model__n_estimators": [90,100,110],
    "model__max_depth": [20,21,22],
    "model__eta": [0.2,0.1,0.095],
    "model__reg_alpha": [20,30],
    "model__reg_lambda": [20,30],
    "model__gamma": [1,10,20],
    'model__colsample_bytree': [0.8,0.9,1],
    'model__subsample': [0.8,0.9,1],
    }

xgb_pipe = Pipeline([('preprocessing',transformer),('model',XGBRegressor(enable_categorical=True,random_state=42))])
search = GridSearchCV(xgb_pipe, 
                            scoring='neg_median_absolute_error', 
                            param_grid=params,
                            cv=cv, n_jobs=-1)
search.fit(x_train, y_train)
print(search.best_score_)
print(search.best_params_)
```

* Jika dilihat dari hasil gridsearch sepertinya XGBoost adalah model terbaik.Namun akan kita cek dulu apakah model ini dapat generalisasi dengan baik pada data yang belum terlihat.

* Terlihat model mampu memprediksi test set dengan MedAE yang hampir sama dengan MedAE pada gridsearch.Sehingga model ini dapat dikatakan mampu memprediksi data yang belum pernah terlihat sebelumnya.

**XGBoost** merupakan ensemble model dengan metode boosting yang berarti menggunakan beberapa model yang biasanya adalah decision tree untuk menghasilkan prediksi.Pada xgboost untuk menghasilkan prediksi yang baik diterapkan technique boosting dimana model weak learner yang biasanya memiliki performa kurang baik akan dilatih kembali dengan memberi perhatian khusus pada prediksi yang masih salah secara iterative, cont:DT_1,DT_2,DT_3,...,DT_n.Jadi cara kerjanya adalah pertama model akan melatih DT_1, model DT_1 ini akan disesuaikan bobotnya pada prediksi yang salah yang nantinya akan dilanjutkan untuk melatih DT_2 dengan bobot yang telah terupdate, proses ini dilakukan untuk mengurangi error yang dihasilkan/kesalahan prediksi yang dilakukan model sebelumnya, yang dimana menghasilkan kinerja overall yang lebih baik, singkatnya adalah model belajar dari kesalahan model sebelumnya.Ilustrasi gambar untuk metric logloss,mae,dsb yang bertujuan menurunkan error.[https://i0.wp.com/neptune.ai/wp-content/uploads/2022/10/Gradient-boosting-LightGBM-vs-XGBoost.png?resize=591%2C431&ssl=1]
